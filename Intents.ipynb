```python
# define our vocabulary

TASKS = ['brush', 'clean', 'clip', 'crush', 'curl', 'cut', 'dig', 'dispense', 
         'drink', 'dust', 'flatten', 'flip', 'funnel', 'grate', 'grind', 'hammer', 
         'hang', 'juice', 'ladle', 'lift', 'lock', 'mash', 'mix', 'open', 'paint', 'peel', 
         'pick up', 'plug in', 'poke', 'pound', 'pour', 'roll', 'saute', 'scoop', 'scrape', 
         'scratch', 'screw', 'scrub', 'shake', 'sharpen', 'sift', 'skewer', 'skim', 'slice', 
         'spray', 'squeegee', 'squeeze', 'stab', 'stir', 
         'straighten', 'strain', 'sweep', 'tenderize', 'till', 'turn on', 'handover']

## 
# objects of interest: spatula, fork, pan, scissor, hammer(?)
# spatula: mix, sift, saute, stir, handover
# fork: stab, skewer, poke, scoop, handover
# pan: pour, pound, scoop, lift, dispense, handover
# scissor: cut, poke, stab, clip, handover
# hammer: crush, pound, mash, handover, flatten
##

actions = [
    "pass", "hand", "give", "grab", "pass to", "hand over to", "give to", "hand over"
]

actions2 = [
    "grasp", "take", "pick up", "grab"
]

descriptors = [
    "blue", "red", "green", "yellow", "black", "white",
    "big", "large", "small", "smaller", "bigger",
    "frontmost", "leftmost", "rightmost", "middle",
    "left", "right"
]

verbs = [
    "use", "pick up", "grab", "try to use", "try using"
]

objects = [
    "mug", "spoon", "fork", "hat", "item", "object", "bottle", "spatula", "hammer", "box",
    "cylinder", "pan", "dish", "bowl", "battery", "container", "scissor", "scissors", "pen",
    "pencil"
]

tasks = [
    "cutting", "pouring", "poking", "eating", "dispensing", "drawing", "usage", "cooking"
]

task2 = [
    "cut", "pour", "hit", "slice", "slash", "smash", "whack", "pound", "stab", "scoop", "lift", "poke", "mix", "saute", "stir"
]

positions = [
    "in front of", "behind the", "to the left of the", "to the right of the"
]

locators = [
    "in the back", "at the front", "on the left", "on the right"
]

fillers = [
    "could you", "can you", "could you please", "can you please", "please", "i want you to"
]

intents = [
    'handover',
    'pour',
    'use',
    'hit',
    'cut',
]
```

```python
import random
import numpy as np
import pandas as pd


dupes = []

def get_intent(task, default = "handover"):
    # spatula: mix, sift, saute, stir, handover
    # fork: stab, skewer, poke, scoop, handover
    # pan: pour, pound, scoop, lift, dispense, handover
    # scissor: cut, poke, stab, clip, handover
    # hammer: crush, pound, mash, handover, flatten
    if task in ["cutting", "cut", "slice", "slash"]:
        return "cut"
    elif task.replace("ing", "") in TASKS:
        return task
    elif task == "cooking":
        return random.choice(['mix', 'sift', 'saute', 'stir'])
    elif task in ["pour", "pouring", "dispensing"]:
        return "pour"
    elif task in ['poking', "hit", "pound", "smash"]:
        return "hit"
    elif task == "usage":
        return 'use'
    else:
        return default

random.seed(42)
for i in range(20000):
    sentence_type = random.randint(0, 6)
    # then permutate randomly based on sentence type
    #     {ACTION} me the {DESCRIPTIOR} {OBJ} -- 0
    #     {ACTION} me the {DESCRIPTIOR} {OBJ} for {TASK} -- 1
    #     {VERB} the {DESCRIPTIOR} {OBJ} to {TASK} -- 2
    #     {ACTION2} the {DESCRIPTIOR} {OBJ} {POSITION} {DESCRIPTIOR} {SECONDARY_OBJ} -- 3
    #     {ACTION2} the {DESCRIPTIOR} {OBJ} {LOCATOR} {LOCATION} -- 4
    #     {TASK2} the {DESCRIPTOR} {SECONDARY_OBJ} WITH {OBJ} -- 5
    #     I WANT THE {OBJ} FOR {TASK} -- 6
    
    FILLER = random.choice(fillers) + " " if random.random() < 0.75 else ""
    ACTION = random.choice(actions)
    ACTION2 = random.choice(actions2)
    DESCRIPTOR = random.choice(descriptors) + " " if random.random() < 1/3 else ""
    DESCRIPTOR2 = random.choice(descriptors) + " " if random.random() < 1/3 else ""
    VERB = random.choice(verbs)
    TASK = random.choice(tasks)
    TASK2 = random.choice(task2)
    POSITION = random.choice(positions)
    LOCATOR = random.choice(locators)
    OBJ = random.choice(objects)
    SECONDARY_OBJ = random.choice(objects)
    
    
    if len(FILLER) > 0 and sentence_type < 6:
        sentence = FILLER
        slot = "O " * (len(FILLER.rstrip().split(" ")))
    else:
        sentence = ""
        slot = ""
        
    
    if sentence_type == 0:
        # {ACTION} me the {DESCRIPTIOR} {OBJ} -- 0
        intent = "handover"
        sentence += f"{ACTION} me the {DESCRIPTOR}{OBJ}"
        slot += "O " * len(ACTION.split(" ")) + "O O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type"
    elif sentence_type == 1:
        # {ACTION} me the {DESCRIPTIOR} {OBJ} for {TASK} -- 1
        intent = get_intent(TASK)
        sentence += f"{ACTION} me the {DESCRIPTOR}{OBJ} for {TASK}"
        slot += "O " * len(ACTION.split(" ")) + "O O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type O O"
    elif sentence_type == 2:
        # {VERB} the {DESCRIPTIOR} {OBJ} to {TASK2} -- 2
        intent = get_intent(TASK2, "use")
        sentence += f"{VERB} the {DESCRIPTOR}{OBJ} to {TASK2}"
        slot += "O " * len(VERB.split(" ")) + "O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type O O"
    elif sentence_type == 3:
        # {ACTION} the {DESCRIPTIOR} {OBJ} {POSITION} {DESCRIPTIOR} {SECONDARY_OBJ} -- 3
        intent = "handover"
        sentence += f"{ACTION2} the {DESCRIPTOR}{OBJ} {POSITION} {DESCRIPTOR2}{SECONDARY_OBJ}"
        slot += "O " * len(ACTION2.split(" ")) + "O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type B-object_position " + "I-object_position " * (len(POSITION.split(" ")) - 1)
        if len(DESCRIPTOR2) > 0:
            slot += "B-secondary_object_descriptor " + "I-secondary_object_descriptor " * (len(DESCRIPTOR2.rstrip().split(" ")) - 1)
        slot += "B-secondary_object"
    elif sentence_type == 4:
        # {ACTION} the {DESCRIPTIOR} {OBJ} {LOCATOR} {LOCATION} -- 4
        intent = "handover"
        sentence += f"{ACTION2} the {DESCRIPTOR}{OBJ} {LOCATOR}"
        slot += "O " * len(ACTION2.split(" ")) + "O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type B-object_position " + "I-object_position " * (len(LOCATOR.split(" ")) - 1)
    elif sentence_type == 5:
        # {TASK2} the {DESCRIPTOR} {SECONDARY_OBJ} WITH {OBJ} -- 5
        intent = get_intent(TASK2, "use")
        sentence += f"{TASK2} the {DESCRIPTOR2}{SECONDARY_OBJ} WITH {DESCRIPTOR}{OBJ}"
        slot += "O " * len(TASK2.split(" ")) + "O "
        if len(DESCRIPTOR2) > 0:
            slot += "B-secondary_object_descriptor " + "I-secondary_object_descriptor " * (len(DESCRIPTOR2.rstrip().split(" ")) - 1)
        slot += "B-secondary_object O "
        if len(DESCRIPTOR) > 0:
            slot += "B-object_descriptor " + "I-object_descriptor " * (len(DESCRIPTOR.rstrip().split(" ")) - 1)
        slot += "B-object_type"
    elif sentence_type == 6:
        # I WANT THE {OBJ} FOR {TASK} -- 6
        intent = get_intent(TASK, "use")
        sentence += f"I WANT THE {OBJ} FOR {TASK}"
        slot += "O O O B_object_type O O"
        
    sentence = sentence.lower()
    dupes.append((sentence, intent, slot))
    
n_arr = np.array(list(set(dupes)))
sets  = np.split(n_arr, [int(0.6*len(n_arr)), int(0.8*len(n_arr))])
dirs = ['train', 'dev', 'test']

for i in range(3):
    main_dir = dirs[i]
    main_bulk = open(f'{main_dir}/seq.in', 'w')
    main_intents = open(f'{main_dir}/label', 'w')
    main_slots = open(f'{main_dir}/seq.out', 'w')
    for j in sets[i]:
        sentence, intent, slot = j
        main_bulk.write(sentence + '\n') 
        main_intents.write(intent + '\n')
        main_slots.write(slot + '\n')
    main_bulk.close()
    main_intents.close()
    main_slots.close()
        
# for i in set(dupes):
#     sentence, intent, slot = i
#     main_bulk.write(sentence + '\n') 
#     main_intents.write(intent + '\n')
#     main_slots.write(slot + '\n')
    

```
